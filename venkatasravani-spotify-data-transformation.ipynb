```python
# Import necessary libraries
import pandas as pd  # Library for data manipulation and analysis
import numpy as np  # Library for numerical operations
import os  # Library for interacting with the operating system
import matplotlib.pyplot as plt  # Library for creating static visualizations
import plotly.express as px  # Library for creating interactive visualizations
import plotly.offline as po  # Library for offline interactive plotting
```

```python
df = pd.read_csv("/kaggle/input/most-streamed-spotify-songs-2024/Most Streamed Spotify Songs 2024.csv", encoding="latin-1")
```

```python
df_missing = pd.DataFrame(df.isnull().sum()).reset_index().rename(columns={'index': 'Track', 0: 'Missing Values'})
```

```python
df_missing.drop(index=df.index[0], axis=0, inplace=True)
df_missing[df_missing['Missing Values']>0].sort_values(by='Missing Values', ascending=False)
```

```python
df = df.drop(['Album Name'],axis='columns')
```

```python
top_Artist = df.Artist.value_counts().head(30)
figure = px.bar(top_Artist, x=top_Artist.index, y=top_Artist.values,title="Popular Artists with their count",labels={'x': 'Artist', 'y': 'Count'})
figure.show()
figure = px.pie(top_Artist, names=top_Artist.index, values=top_Artist.values,title="Popular Artists with their count")
figure.show()
```
